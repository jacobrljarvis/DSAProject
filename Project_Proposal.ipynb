##NBA Player Stats 2023-2024 Season
Kaggle - "2023-2024 NBA Player Stats"

https://www.kaggle.com/datasets/vivovinco/2023-2024-nba-player-stats/data

Format: CSV


###Contents
The dataset contains NBA player statistics from the 2023-2024 regular season. It includes traditional states, shooting percentages and advanced metrics for all NBA players who appeared in games during the season.


###Attributes (List that can be chosen from):

* Games (G), Games Started (GS), Minutes Played (MP)
* Field Goals (FG, FGA, FG%)
* 3-Point Shots (3P, 3PA, 3P%)
* Free Throws (FT, FTA, FT%)
* Rebounds (ORB, DRB, TRB)
* Assists (AST)
* Steals (STL)
* Blocks (BLK)
* Turnovers (TOV)
* Personal Fouls (PF)
* Points (PTS)
* Per-game versions of above stats (PPG, RPG, APG)
* Player Efficiency Rating (PER)
* True Shooting Percentage (TS%)
* Usage Rates (USG%)
* Position (PG, SG, SF, PF, C) -- **Target Value**



Anticipated Challenges
* Many players have more than one position
  * Solution: Choose their primary position or treat as a seperate class
* Guards (PG/SG) may be overrepresented compared to big men
  * Solution: Use stratified test/train splits or use class weights in classifiers
* Basketball has mordernized away from certain positions doing only certain tasks, stats based classifying may not list certain superstars that transcend traditional positional stats
  * Solution: Creates certain outliers that show "positionless" players, interesting for analysis
* Missing Values, Players who didn't attempt 3-pointers or free throws may have NaN for percentages
  * Solution: Imputer 0% or league averages


Feature Selection Strategy
*   Focus on per-game stats rather than totals (eliminates playing time bias)
*   Reduces to around 12-15 core metrics



###Potential Impact of Results
1. Player Development & Scouting
  * Identify players whose statistics doesn't match their listed position
2. Lineup Optimization
  * Help coaches build balanced lineups by understanding players' statistical profiles
3. Understanding Basketball Evolution
  * Quantify the trend toward "positionless basketball"
  * Track how the game has evolved
